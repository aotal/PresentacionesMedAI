# Clasificación de los datos

## Por tipos de datos

-   **Datos estructurados**: Datos organizados en tablas o matrices, como datos numéricos en
una hoja de cálculo.
-   **Datos no estructurados**: Datos que no tienen una forma predefinida de organización,
como texto libre, imágenes, audio o video.
-   **Datos semiestructurados**: Datos que tienen alguna estructura pero no encajan perfectamente
en un formato tabular, como archivos XML o JSON.
